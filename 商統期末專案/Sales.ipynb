```r
sales$CustCode <- as.factor(sales$CustCode)
sales$客戶名稱 <- as.factor(sales$客戶名稱)
sales$Brand <- as.factor(sales$Brand)
sales$BrandName <- as.factor(sales$BrandName)
sales$CustCode <- as.factor(sales$CustCode)
sales$ProdtCode <- as.factor(sales$ProdtCode)
sales$類型 <- as.factor(sales$類型)
sales$主型號_楦 <- as.factor(sales$主型號_楦)
sales$性別 <- as.factor(sales$性別)
sales$顏色 <- as.factor(sales$顏色)
sales$尺寸 <- as.character(sales$尺寸)
sales$主型號 <- as.factor(sales$主型號)

str(sales)
```

### 性別、尺寸、顏色 vs. 銷量

```r
by.gender <- sales %>%
    group_by(性別)  %>%
    summarise(amountSold = sum(數量), avgSold = mean(數量),.groups='drop')  #不能加$字號

by.gender
```

```r
par(family="Songti TC")
par(pin = c(5,3))
barplot(by.gender$amountSold ~ by.gender$性別, xlab="Gender", ylab="amount Sold")
```

### 整理尺寸

```r
#newsize <- data.frame()
for (i in c(1: nrow(sales))){
    if (grepl(pattern = "-", x = sales$尺寸[i]) == TRUE){
        tmp <- unlist(strsplit(toString(sales[i,18]), split='-', fixed=TRUE))[2]
    }else{tmp <- as.character(sales[i,18])}
    sales$new尺寸[i] <- tmp
}
```

```r
sales$new尺寸 <- as.factor(sales$new尺寸)

by.size <- sales%>%
  group_by(new尺寸)%>%
  summarise(amount=sum(數量),avgSold = mean(數量), .groups='drop') %>%
  arrange(desc(amount))

by.size
```

```r
par(pin = c(5,3))
barplot(by.size$amount ~ by.size$new尺寸, xlab="Size", ylab="amount Sold")
```

```r
by.color <- sales%>%
  group_by(顏色)%>%
  summarise(amount=sum(數量),avgSold = mean(數量), .groups='drop') %>%
  arrange(desc(amount))

by.color
```
