## 分群討論：鞋類品項、非鞋類品項

```r
## 分群：非鞋類
non_shoes_list = c(0, 'F', 'L', 'M', 'S', 'XL', 'XS', 'XXL', 'XXS')
non_shoes <- sales %>%
  filter(new尺寸 %in% non_shoes_list & 金額>0)
```

```r
sales_over10000 <- sales %>%
  filter(金額 > 10000 & !(new尺寸 %in% non_shoes_list))
sales_below10000 <- sales %>%
  filter(金額<=10000 & !(new尺寸 %in% non_shoes_list) & 金額 > 0)
```

### Model 2：log (金額)

```r
attach(sales_below10000)
test_mod2 <- lm(log(金額)~ 客戶名稱+Brand+打折比率+類型+性別+顏色+new尺寸+主型號, sales_below10000)
anova(test_mod2)
```

```r
summary(test_mod2)
```

```r
par(mfrow = c(2, 2))
plot(test_mod2)
```

根據上次EDA觀察，不同尺寸、鞋款性別、Brand下的商品打折比率會有所不同，因此嘗試加入打折比率與其他類別變數做interaction，並觀察其結果如何。

### Model 3：加入交互作用項 (w/ 打折比率)

```r
attach(sales_below10000)
test_mod3 <- lm(log(金額)~ 打折比率*(客戶名稱+Brand+類型+性別+顏色+new尺寸+主型號), sales_below10000)
anova(test_mod3)
```

```r
summary(test_mod3)
```

```r
par(mfrow = c(2, 2))
plot(test_mod3)
```

### Model 4：去除outlier

```r
cook.mod3 <- cooks.distance(test_mod3)
plot(cook.mod3, xlab="sales", ylab="cooks distance")
```
