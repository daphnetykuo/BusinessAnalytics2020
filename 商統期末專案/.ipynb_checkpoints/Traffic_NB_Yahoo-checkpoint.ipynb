### EDA

<b> 工作階段

```r
attach(NBtraffic)

NBtraff.ss.summary <- NBtraffic %>%
  group_by(month(日期)) %>%
  summarize("最大工作階段"= max(工作階段), "最小工作階段"=min(工作階段), "月均工作階段"=mean(工作階段), "月總工作階段"=sum(工作階段)) %>%
  ungroup()

#par(pin = c(3,1.5))
#par(mfrow = c(2, 1))
#plot(工作階段~日期, xlab="Date", ylab="Sessions")
boxplot(工作階段 ~ month(日期), xlab="Month", ylab="Sessions")
```

<b> 跳出率 (Bounce Rate)

```r
attach(NBtraffic)

NBtraff.br.summary <- NBtraffic %>%
  group_by(month(日期)) %>%
  summarize("最高跳出率"= max(跳出率), "最低跳出率"=min(跳出率), "月均跳出率"=mean(跳出率)) %>%
  ungroup()

#par(pin = c(3,1.5))
#par(mfrow = c(2, 1))
#plot(工作階段~日期, xlab="Date", ylab="Sessions")
boxplot(跳出率 ~ month(日期), xlab="Month", ylab="Bounce Rate")
```

<b> 單次工作頁數 (PageViews/Sessions)

```r
attach(NBtraffic)

NBtraff.pvs.summary <- NBtraffic %>%
  group_by(month(日期)) %>%
  summarize("最高單次工作階段頁數"= max(單次工作階段頁數), "最低單次工作階段頁數"=min(單次工作階段頁數), "月均單次工作階段頁數"=mean(單次工作階段頁數), "月總單次工作階段頁數"=sum(單次工作階段頁數)) %>%
  ungroup()

boxplot(單次工作階段頁數 ~ month(日期), xlab="Month", ylab="Page Views/ Sessions")
```

```r
print("NB Session")
NBtraff.ss.summary %>%
  arrange(desc(月總工作階段))
print("NB Bounce Rate")
NBtraff.br.summary %>%
  arrange(desc(月均跳出率))
print("NB Page Views/ Session")
NBtraff.pvs.summary %>%
  arrange(desc(月均單次工作階段頁數))
```

NB流量月份整理
  1. 四月單次工作階段頁數最高、跳出率居中，但月總工作階段數最低。可以觀察當月業績的高低判斷。
  2. 八月月總工作階段數最高(訪客最多)、跳出率居中高、單次工作階段頁數偏低。可與四月比較業績高低。
  3. 七月跳出率最低、單次工作階段頁數第二。可觀察當月業績是否真的比較高。
  
目的：檢驗網站流量與業績是否正相關。若正相關，再檢驗當月打折比率情況如何or當月銷售商品以和為主。

# 通路業績情況

```r
sales <- read_excel("/Users/daphne/Desktop/Business_Analytics/FinalProject/NBdata/newSales.xlsx")
head(sales)
```

```r
sales$CustCode <- as.factor(sales$CustCode)
sales$客戶名稱 <- as.factor(sales$客戶名稱)
sales$Brand <- as.factor(sales$Brand)
sales$BrandName <- as.factor(sales$BrandName)
sales$ProdtCode <- as.factor(sales$ProdtCode)
sales$類型 <- as.factor(sales$類型)
sales$主型號.楦 <- as.factor(sales$主型號.楦)
sales$性別 <- as.factor(sales$性別)
sales$顏色 <- as.factor(sales$顏色)
sales$new尺寸 <- as.factor(sales$new尺寸)
sales$主型號 <- as.factor(sales$主型號)
sales$簡化類型 <- as.factor(sales$簡化類型)
sales$分類 <- as.factor(sales$分類)
sales$SlipDate <- ymd(sales$SlipDate)

str(sales)
```

## NB業績

```r
NBsales.summary <- sales %>%
  filter(客戶名稱 == "NB") %>%
  group_by(month(SlipDate)) %>%
  summarize("月均銷售額"=mean(金額), "月總銷售額"=sum(金額), "月均銷售量"=mean(數量), "月總銷售量"=sum(數量), "月總實售單價"=sum(實售單價), "月總建議售價"= sum(建議售價)) %>%
  ungroup()

NBsales.summary <- NBsales.summary %>%
  mutate("月打折比率" = 月總實售單價/月總建議售價)

names(NBsales.summary)[names(NBsales.summary) == "month(SlipDate)"] <- "month(日期)"

NBsales.summary %>%
  arrange(desc(月總銷售額))

NBtraff.ss.summary %>%
  arrange(desc(月總工作階段))
```

```r
## 流量資料（週）
NBtraff.ss.w.summary <- NBtraffic %>%
  group_by(week(日期)) %>%
  summarize("最大工作階段"= max(工作階段), "最小工作階段"=min(工作階段), "週均工作階段"=mean(工作階段), "週總工作階段"=sum(工作階段)) %>%
  ungroup()

NBtraff.br.w.summary <- NBtraffic %>%
  group_by(week(日期)) %>%
  summarize("最高跳出率"= max(跳出率), "最低跳出率"=min(跳出率), "週均跳出率"=mean(跳出率)) %>%
  ungroup()

NBtraff.pvs.w.summary <- NBtraffic %>%
  group_by(week(日期)) %>%
  summarize("最高單次工作階段頁數"= max(單次工作階段頁數), "最低單次工作階段頁數"=min(單次工作階段頁數), "週均單次工作階段頁數"=mean(單次工作階段頁數), "週總單次工作階段頁數"=sum(單次工作階段頁數)) %>%
  ungroup()

## 業績資料（週）
NBsales.w.summary <- sales %>%
  filter(客戶名稱 == "NB") %>%
  group_by(week(SlipDate)) %>%
  summarize("週均銷售額"=mean(金額), "週總銷售額"=sum(金額), "週均銷售量"=mean(數量), "週總銷售量"=sum(數量), "週總實售單價"=sum(實售單價), "週總建議售價"= sum(建議售價)) %>%
  ungroup() %>%
  mutate("週打折比率" = 週總實售單價/週總建議售價)

names(NBsales.w.summary)[names(NBsales.w.summary) == "week(SlipDate)"] <- "week(日期)"

## 合併表格
NBw.dt <- merge(x = NBsales.w.summary, y = NBtraff.ss.w.summary, by = "week(日期)")
NBw.dt <- merge(x = NBw.dt, y = NBtraff.br.w.summary, by = "week(日期)")
NBw.dt <- merge(x = NBw.dt, y = NBtraff.pvs.w.summary, by = "week(日期)")
```

```r
head(NBw.dt)
```

```r
colnames(NBw.dt)  # 業績、session、bounce rate、page view per session
```

```r
par(family="Songti TC")
plot(NBw.dt[,c(1,3,5,8,12,15,18,19)])
```

銷售量和銷售額略呈現右偏（取log）；週總單次工作階段頁數也要取log

<b> Y：週總工作階段 or 週總單次工作階段頁數
<br>    ➤ 目的：看打折比率對瀏覽數量是否有影響

```r
plot(週總工作階段 ~ 週均跳出率, xlab="mean Bounce Rate", ylab="weekly total Session", data=NBw.dt)
#curve(lm)
```

```r
attach(NBw.dt)
nb.mod1.1 <- lm(log(週總工作階段) ~ 週打折比率+週均跳出率+log(週總單次工作階段頁數))
summary(nb.mod1.1)

par(mfrow = c(2, 2))
plot(nb.mod1.1)
```

```r
週均跳出率_sqr <- (NBw.dt$週均跳出率^2)
nb.mod1.2 <- lm(log(週總工作階段) ~ 週打折比率+週均跳出率+週均跳出率_sqr+log(週總單次工作階段頁數), data=NBw.dt)
summary(nb.mod1.2)

par(mfrow = c(2, 2))
plot(nb.mod1.2)
```

```r
NBw.dta <- NBw.dt[-c(39),]
```

```r
attach(NBw.dta)
週均跳出率_sqra <- (NBw.dta$週均跳出率^2)
nb.mod1.3 <- lm(log(週總工作階段) ~ 週打折比率+週均跳出率+週均跳出率_sqra+log(週總單次工作階段頁數),data=NBw.dta)
summary(nb.mod1.3)

par(mfrow = c(2, 2))
plot(nb.mod1.3)
```
