```r
library(readxl)
library(dplyr)
library(ggplot2)
```

```r
sales <- read_excel("/Users/daphne/Desktop/Business_Analytics/FinalProject/NBdata/業績_新增商品資料欄位.xlsx")
sales
```

```r
sales$客戶名稱 <- as.character(sales$客戶名稱)
sales$客戶名稱[sales$客戶名稱 == "紐巴倫(蝦皮)"] <- "Shopee"
sales$客戶名稱[sales$客戶名稱 == "紐巴倫(樂天)"] <- "Lotte"
sales$客戶名稱[sales$客戶名稱 == "紐巴倫官網"] <- "NB"
sales$客戶名稱[sales$客戶名稱 == "MOMO紐巴倫"] <- "MOMO"
sales$客戶名稱[sales$客戶名稱 == "YAHOO紐巴倫"] <- "YAHOO"
sales$CustCode <- as.factor(sales$CustCode)
sales$客戶名稱 <- as.factor(sales$客戶名稱)
sales$Brand <- as.factor(sales$Brand)
sales$BrandName <- as.factor(sales$BrandName)
sales$ProdtCode <- as.factor(sales$ProdtCode)
sales$類型 <- as.factor(sales$類型)
sales$主型號.楦 <- as.factor(sales$主型號.楦)
sales$性別 <- as.factor(sales$性別)
sales$顏色 <- as.factor(sales$顏色)
sales$尺寸 <- as.character(sales$尺寸)
sales$主型號 <- as.factor(sales$主型號)
```

```r
#sales$尺寸 <- as.character(sales$尺寸)
for (i in c(1: nrow(sales))){
    if (grepl(pattern = "-", x = sales$尺寸[i]) == TRUE){
        tmp <- unlist(strsplit(toString(sales[i,18]), split='-', fixed=TRUE))[2]
    }else{tmp <- as.character(sales[i,18])}
    sales$new尺寸[i] <- tmp
}

sales$new尺寸 <- as.factor(sales$new尺寸)
```

```r
str(sales)
```

### 分產品：鞋子、衣服、其他

```r
## 分群：鞋類
shoes <- sales %>%
  filter(金額 > 0 & !(new尺寸 %in% non_shoes_list))
'sales_over10000 <- sales %>%
  filter(金額 > 10000 & !(new尺寸 %in% non_shoes_list))
sales_below10000 <- sales %>%
  filter(金額<=10000 & !(new尺寸 %in% non_shoes_list) & 金額 > 0)'
```

```r
## 分群：非鞋類
non_shoes_list = c(0, 'F', 'L', 'M', 'S', 'XL', 'XS', 'XXL', 'XXS')
non_shoes <- sales %>%
  filter(new尺寸 %in% non_shoes_list & 金額>0)
```

```r
## 分群：非鞋類--衣服
clothes <- non_shoes %>%
  filter(new尺寸 != 0)

## 分群：非鞋類--非衣服
non_clothes <- non_shoes %>%
  filter(new尺寸 == 0)
```

## 1. 衣服品項

```r
# 衣服類金額為右偏分配，Y取對數
attach(clothes)
cl.mod1 <- lm(log(金額) ~ 客戶名稱+Brand+打折比率+性別+顏色+new尺寸, data=clothes)
```

```r
anova(cl.mod1)
```

```r
summary(cl.mod1)
```

```r
par(mfrow = c(2, 2))
plot(cl.mod1)
```

```r
clothes[49,]
```

```r
res.cl1 <- rstandard(cl.mod1)
#hist(res.cl1)
print(res.cl1[res.cl1<(-3)|res.cl1>3])
```

```r
outlier <- c(res.cl1[res.cl1<(-3)|res.cl1>3])
```

### model 2: 去除尺寸

```r
cl.mod2 <- lm(log(金額) ~ 客戶名稱+Brand+打折比率+性別+顏色, data=clothes)
anova(cl.mod2)
```

```r
summary(cl.mod2)
```

```r
res.cl2 <- rstandard(cl.mod2)
res.cl2[res.cl2<(-5)|res.cl2>5] # 193,471,622, 627,760,925,940,993
```

```r
clothes.a <- clothes[-c(193,471,622, 627,760,925,940,993),]
attach(clothes.a)
cl.mod3 <- lm(log(金額) ~ 客戶名稱+Brand+主型號+數量+打折比率+性別+顏色+new尺寸, data=clothes.a)
anova(cl.mod3)
summary(cl.mod3)
```

```r
par(mfrow = c(2, 2))
plot(cl.mod3)
```

加入「類型」變數時，model residual plot會出現一段空白斷層(approx. x=5~6)，但R-square可以到57%(上升27個百分比)
